```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Сгенерируем выборку объёма $1000$ для экспоненциального распределения:

```python
exp_rv = sts.expon()
sample = exp_rv.rvs(1000)
```

Построим гистограмму и поверх неё нарисуем теоретическую плотность распределения случайной величины:

```python
plt.hist(sample, bins=20, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
x = np.linspace(-4, 4, 100)
pdf = exp_rv.pdf(x)
plt.plot(x, pdf)
```

Для трёх значений $n$ $(n = 5, n = 10, n = 50)$ сгенерируем 1000 выборок объёма $n$

$1)$ $n = 5$:

```python
sample1 = []
for i in range(1000):
    sum = 0
    test = exp_rv.rvs(5)
    selection = np.array(test)
    for j in range(len(selection)):
        sum += selection[j]
    average = sum/5
    sample1.append(average)
plt.hist(sample1, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
mu = exp_rv.expect()
sigma = np.sqrt(exp_rv.var()/5)
norm_rv = sts.norm(mu, sigma)
x = np.linspace(-1, 3, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical cdf')
```

$2)$ $n = 10$:

```python
sample2 = []
for i in range(1000):
    sum = 0
    test = exp_rv.rvs(10)
    selection = np.array(test)
    for j in range(len(selection)):
        sum += selection[j]
    average = sum/10
    sample2.append(average)
plt.hist(sample2, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
mu = exp_rv.expect()
sigma = np.sqrt(exp_rv.var()/10)
norm_rv = sts.norm(mu, sigma)
x = np.linspace(-1, 3, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical cdf')
```

$3)$ $n = 50$:

```python
sample3 = []
for i in range(1000):
    sum = 0
    test = exp_rv.rvs(50)
    selection = np.array(test)
    for j in range(len(selection)):
        sum += selection[j]
    average = sum/50
    sample3.append(average)
plt.hist(sample3, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
mu = exp_rv.expect()
sigma = np.sqrt(exp_rv.var()/50)
norm_rv = sts.norm(mu, sigma)
x = np.linspace(-4, 4, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical cdf')
```

Параметры нормального распределения были получены следующим образом:
$1)$ Математическое ожидание экспоненциального распределения равно $\lambda^{-1}$, в нашем случае $\lambda = 1$, следовательно $\mu = 1$. $2)$ Дисперсия экспоненциального распределения равна $\lambda^{-2}$, поэтому среднеквадратичное отклонение нормального распределения будет равно $\sqrt\frac{\lambda^{-2}}{n}$ и для $n = 5, 10, 50$ получаем соответствующие отклонения.

Из графиков можно сделать вывод, что при увеличении $n$ точность аппроксимации распределения выборочных средних улучшается, т.к. при увеличении $n$ гистограмма всё больше соответствует нормальному распределению.
